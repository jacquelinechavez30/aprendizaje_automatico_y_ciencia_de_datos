a) Cargar el dataset con URL indicada, en un DataFrame, mostrar cuantas filas y columnas tiene el dataset, también mostrar todas las columnas y por cada columna mostrar el nombre de la columna según el siguiente formato, cambiar el tipo de dato con un identificador en Español: int64 por entero64, object por "contenedor general", y remover la columna “Non-Null Count (5%)".

```python
import pandas as pd
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data")
print(df.info)#No tienen nombres las columnas se los asigno
columnas_nombres=["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation" "relationship", "race", "gender", "capital-gain", "capital-loss", "hours-per-week", "native-country", "income"]
df=pd.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/adult/adult.data", names=columnas_nombres, skipinitialspace=True)
print("Monstrando la primer fila(solo para visualizar los cambios) del data frame con los nombres de las columnas asignados")
print(df.columns)
#Cuantas filas y columnas tiene el data frame
print("Numero de Filas:", df.shape[0], "Numero de Columnas:", df.shape[1])
#Mostrar las columnas y cambiar el tipo de deato
print("Tipos de datos de cada columna y numero de columna N")
#creando una serie apartir del dtypes del data frame y reemplazando  el nombre de los tipos de datos
df_dtypes=df.dtypes.replace({"object": "Contenedor General", "int64": "Entero64"})
cantidad=df.count().values 
#creando un data frame apartir de la serie creada y agregando una columna con el numero de columna
formato = pd.DataFrame({
    ' No ': range(len(df_dtypes)),  
    'Columna': df_dtypes.index, 
    'Non-Null Count':  cantidad.astype(str) + " non-null",
    'Tipo': df_dtypes.values 
})
#Mostrando el data frame con el formato  solicitado en la guia de ejercicios
print("Mostrando el formato solicitado en la guia de ejercicios")
print(formato.to_string(index=False))
#Eliminar la columna  que dice la guia 
formato= formato.drop(columns=['Non-Null Count'])
print("Mostrando el formato solicitado en la guia de ejercicios sin la columna Non-Null Count")
print(formato.to_string(index=False))






```

```python
import io
# Usar io.StringIO para capturar la salida de df.info()
buffer = io.StringIO()
df.info(buf=buffer)
info_string = buffer.getvalue()
print(info_string)
#Dividir la cadena en columnas y filas
info_lines = info_string.split('\n')
info_data = [line.split() for line in info_lines if line]
#Crear un data frame con la informacion obtenida
info_df = pd.DataFrame(info_data[3:-1], columns=info_data[3])
print(info_df)
#eliminar las columnas Non-Null Count
info_df = info_df.drop(columns=['Non-Null', 'Count'])
#eliminar las primeras dos filas
info_df = info_df.iloc[2:]
print("Columnas del data frame info_df:")
print(info_df.to_string(index=False))
print(info_df.columns)
```

c) Utilizando una sola orden, obtener información de Porcentaje de ingresos >50K por país.


```python
#Forma 1
porcentaje_ingreso_por_pais_mayor_50k = df.groupby('native-country')['income'].apply(
    lambda x: (x == '>50K').mean() * 100
)
#Forma 2
prueba=(df['income'] == '>50K').groupby(df['native-country']).mean() * 100

#Mostrando los resultados
print("Porcentaje  de 50K por país:" , porcentaje_ingreso_por_pais_mayor_50k)
print("Porcentaje de  50K por país (prueba):" , prueba)

#cuantas filas hay por pais
filas_por_pais = df['native-country'].value_counts()
print("Número de filas por país:", filas_por_pais)
```
